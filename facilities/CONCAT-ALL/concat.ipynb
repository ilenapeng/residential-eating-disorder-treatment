```python
import pandas as pd
```

```python
df_samhsa = pd.read_csv('../samhsa/clean/samhsa-treatment-data-subset-cols.csv')
df_mn = pd.read_csv('../monte-nido/monte-nido-for-concat.csv')
df_carf = pd.read_csv('../carf/carf-accredited-for-concat.csv')
```

```python
# add a source column for all of them to know where the data came from
df_samhsa['data_source'] = 'Substance Abuse and Mental Health Services Administration'
df_mn['data_source'] = 'Monte Nido & Affiliates'
df_carf['data_source'] = 'Commission on Accreditation of Rehabilitation Facilities'
```

```python
df_mn.head(2)

# rename org to name
# rename street to street1

df_mn = df_mn.rename(columns = {'org': 'name1', 'street': 'street1'})

# remove address, program_type, program_type_specific columns
df_mn = df_mn.drop(columns=['address', 'program_type', 'program_type_specific'])
```

```python
df_carf.head(2)

# rename address to street
# rename name to name1
# rename company to name2
df_carf = df_carf.rename(columns = {'address': 'street1', 'name': 'name1', 'company': 'name2'})

# remove address_pt2, link, type columns
df_carf = df_carf.drop(columns=['address_pt2', 'link', 'type'])
```

```python
df_stacked = pd.concat([df_samhsa, df_mn, df_carf]) 
df_stacked.head()
```

```python
df_stacked.to_csv('compiled-data.csv', index=False)
```
